```python
# import py_entitymatching package
import py_entitymatching as em
import os
```

```python
# specify filepaths for tables A and B. 
path_A = 'tableA.csv'
path_B = 'tableB.csv'
```

```python
# read table A; table A has 'ID' as the key attribute
A = em.read_csv_metadata(path_A, key='id')

# read table B; table B has 'ID' as the key attribute
B = em.read_csv_metadata(path_B, key='id')
```

```python
A # print table A to ensure correctly loaded
```

```python
B # print table A to ensure correctly loaded
```

```python
# Attribute equivalence blocking: block A, B  using year
ab = em.AttrEquivalenceBlocker()
C0 = ab.block_tables(A, B, 'year', 'year',
                    l_output_attrs=['name', 'year'], 
                    r_output_attrs=['name', 'year'])
C0
```

```python
B.loc[B['year'] == 1988]
```

```python
# Overlap blocking over candidata set (C0)
ob = em.OverlapBlocker()

# block using name
C1 = ob.block_candset(C0, 'name', 'name', rem_stop_words=True, word_level=True, overlap_size=1)
```

```python
C1 # Our final table of canidates to eventually do matching on. 
# This is a combination of blocking first on year and then using name
```

```python
# now debug the blocker output
dbg = em.debug_blocker(C1, A, B, output_size=125)
```

```python
dbg # Look at the debug table for things that match. If we find things that match, that indicates that our blocker 
# is dropping things that it shouldn't.
```

```python
dbg[['ltable_name', 'rtable_name']] # Look for name matches in the 
# debug table. (We're doing this in addition to the above full debug
# table print out since the rtable_name is hidden away in the collapse
# on the above call.)

# Looking we see no matches so our blocking seems pretty good
```

```python
# export C0 (the canidate set from just the year blocking)
em.to_csv_metadata(C0, './candidate_set_C0_v2.csv')
```

```python
# export C1 (the canidate set from blocking on year and name. This is our final set)
em.to_csv_metadata(C1, './candidate_set_C1_v2.csv')
```

```python
# read back C1 to ensure it saved
C = em.read_csv_metadata('candidate_set_C1_v2.csv', ltable=A, rtable=B)
```
